```python
class mesh_shape:
    def __init__(self,path_to_mesh_shp):
        self.mesh=path_to_mesh_shp
    def get_gridprops_disv(self):
        #Method to get vertices, cell2d, ncpl
        def reorder_coords(coords):
            # coords = mapping(sh_polygon)['coordinates'][0]
            re_coords=[]
            for element in coords:
                re_coords.append(element)
    
            re_coords=re_coords[::-1] # clockwise ordering
            #re_coords.pop()
            return re_coords
		#Save clipped polygons to shapefile o geojson check later
        vorGdf = gpd.read_file(self.mesh)

        g = [i for i in vorGdf.geometry]

        all_coords = []
        for poligono in g: # for first feature/row
            vertex=reorder_coords(mapping(poligono)['coordinates'][0]) #clockwise reorder 
            for element in vertex:
                all_coords.append(element)

        all_coords=list(OrderedDict.fromkeys(all_coords))
        #creating a topology of the order of vertices            
        topology = pd.DataFrame({'ID':np.arange(0,len(all_coords)),'coords':all_coords})

        #dictionary for mapping later each coords with its respective ID
        mapeo={}
        for i,element in enumerate(topology.coords):
            mapeo[element]=topology.ID[i]

        cell2d=[]
        coords_xy_yc=[]
        for element,poligono in enumerate(g):
    
            local_vertexs=reorder_coords(mapping(poligono)['coordinates'][0])
            vertices=[element,poligono.centroid.x,poligono.centroid.y,len(local_vertexs)]
            coords_xy_yc.append((poligono.centroid.x,poligono.centroid.y))
		    
            for vertx in local_vertexs:
                vertices.append(mapeo[vertx])
            cell2d.append(vertices)


        vertices_unfix=topology.to_numpy()
		#vertices=np.zeros((vertices_unfix.shape[0],3))
        vertices=[]
        for k in range(vertices_unfix.shape[0]):
            point=vertices_unfix[k,]
            #vertices[k,]=np.array([point[0],point[1][0],point[1][1]])
            vertices.append([point[0],point[1][0],point[1][1]])

        ncpl=len(cell2d)
        nvert=len(vertices)

        return {'ncpl':ncpl,'cell2d':cell2d,'nvert':nvert,'vertices':vertices,'centroids':coords_xy_yc}    
```

```python
f=open('../Out/basin/disvDict.json')
data = json.load(f)
```

```python
cell2d = data['cell2dArrays']
vertices = data['indexedVerticesList']
ncpl = data['NCPL']
nvert = data['NVERT']
centroids=gridprops['centroids']#data['centroids']
```

# Defininig geometry 

```python
#Extact dem values for each centroid of the voronois
src = rasterio.open('../In/basin/rst/DEM_200b.tif')
elevation=[x for x in src.sample(centroids)]
```

```python
nlay = 5

mtop=np.array([elev[0] for i,elev in enumerate(elevation)])
zbot=np.zeros((nlay,ncpl))


AcuifInf_Bottom = 3000
zbot[0,] = mtop - 30
zbot[1,] = AcuifInf_Bottom + (0.85 * (mtop - AcuifInf_Bottom))
zbot[2,] = AcuifInf_Bottom + (0.70 * (mtop - AcuifInf_Bottom))
zbot[3,] = AcuifInf_Bottom + (0.50 * (mtop - AcuifInf_Bottom))
zbot[4,] = AcuifInf_Bottom
```

# model mf6

```python
# create simulation
model_name = 'test_mesh'
model_ws = 'Resultados'
exe_name = '../exe/mf6.exe'

sim = flopy.mf6.MFSimulation(sim_name=model_name, version='mf6', exe_name=exe_name,sim_ws=model_ws)
```

```python
# create tdis package
tdis_rc = [(1.0, 1, 1.0)]
nper=len(tdis_rc)
tdis = flopy.mf6.ModflowTdis(sim, nper=nper, time_units='seconds',perioddata=tdis_rc)
```

```python
# create iterative model solution and register the gwf model with it
ims = flopy.mf6.ModflowIms(sim, linear_acceleration='BICGSTAB')
```

```python
# create gwf model
gwf = flopy.mf6.ModflowGwf(sim, modelname=model_name, save_flows=True, newtonoptions=['under_relaxation'])
```

```python
# disv
disv = flopy.mf6.ModflowGwfdisv(gwf, nlay=nlay, ncpl=ncpl,top=mtop, botm=zbot,nvert=nvert, vertices=vertices,cell2d=cell2d)
```

```python
ic = flopy.mf6.ModflowGwfic(gwf, strt=np.stack([mtop for i in range(nlay)]))
```

```python
Kx =[4E-4,5E-6,1E-6,9E-7,5E-7]
icelltype = [1,1,0,0,0]
npf = flopy.mf6.ModflowGwfnpf(gwf,save_specific_discharge=True,icelltype=icelltype,k=Kx)
```

```python
rchr = 0.15/365/86400
rch = flopy.mf6.ModflowGwfrcha(gwf, recharge=rchr)
evtr = 1.2/365/86400
evt = flopy.mf6.ModflowGwfevta(gwf,ievt=1,surface=mtop,rate=evtr,depth=1.0)
```

```python
ix2 = GridIntersect(tgr)
```

```python
rios=gpd.read_file('../In/basin/shps/rios.shp')
list_rivers=[]
for i in range(rios.shape[0]):
    
    list_rivers.append(rios['geometry'].loc[i])
    
mls = MultiLineString(lines=list_rivers)
#intersec rivers with our grid
result=ix2.intersect(mls)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
pmv = fplot.PlotMapView(ax=ax, modelgrid=tgr)
pmv.plot_grid()
ix2.plot_linestring(result, ax=ax, lw=3)

for cellid in result.cellids:
    (h2,) = ax.plot(
        tgr.xcellcenters[cellid],
        tgr.ycellcenters[cellid],
        "kx",
        label="centroids of intersected gridcells",
    )

ax.legend([h2], [i.get_label() for i in [h2]], loc="best");
```

```python
#stress_period_data : [cellid, elev, cond, aux, boundname]
drain_list = []
for i in result.cellids:
    drain_list.append([0,i,mtop[i],0.001])
drain_spd = {0:drain_list}
```

```python
drn = flopy.mf6.ModflowGwfdrn(gwf,stress_period_data=drain_spd)
```
